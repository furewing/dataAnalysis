## <a name="aaa">데이터산업 시장규모</a>

### 1. 파일의 데이터 읽어오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

filename = './데이터산업_시장규모_20240429194930.csv'
df = pd.read_csv(filename)

# 업종구분을 인덱스로 설정
df = df.set_index('데이터산업별(1)')

# 데이터 시각화
df.plot(kind='bar', figsize=(12, 6))
plt.xlabel('데이터산업별', fontsize=12)
plt.ylabel('시장규모', fontsize=12)
plt.title('데이터 산업별 시장규모', fontsize=16)
plt.xticks(rotation=10)  # x축 라벨 회전
plt.legend(title='연도', loc='upper left')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

filename = './데이터_솔루션_업종별_매출_비중_20240429211140.csv'
df = pd.read_csv(filename)
df
team_df = df[df['중분류별(1)']=='데이터 수집']
team_df

#2022년 데이터만 사용
values = team_df['2016'].values
labels = team_df['업종별(1)'].values

# 원그래프 생성
fig, ax = plt.subplots(figsize=(3, 3))
ax.pie(values, labels=labels, autopct='%1.1f%%')
ax.axis('equal')  # 원그래프 비율 유지
ax.set_title('데이터_솔루션_업종별_매출_비중', loc='left', pad=20, fontsize=10, fontweight='bold', color='blue')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = ['데이터 수집', '데이터 설계', 'DBMS', '데이터 관리', '데이터 품질', '데이터 분석']#, '데이터 플랫폼']

filename = './데이터_솔루션_업종별_매출_비중_20240429211140.csv'
df = pd.read_csv(filename)

# 그래프 공간 설정
fig, ax = plt.subplots(2, 3, figsize=(15, 10))
plt.subplots_adjust(wspace=0.4, hspace=0.4)  # 그래프 간격 조정

for idx, value in enumerate(data):
  team_df = df[df['중분류별(1)']==value]
  values = team_df['2016'].values
  labels = team_df['업종별(1)'].values
  # 원그래프 생성
  x = idx // 3
  y = idx % 3
  ax[x,y].pie(values, labels=labels, autopct='%1.1f%%')
  ax[x,y].axis('equal')  # 원그래프 비율 유지
  ax[x,y].set_title(data[idx], loc='center', fontsize=10, fontweight='bold', color='blue')
  #ax[x,y].set_title('데이터_솔루션_업종별_매출_비중', loc='center', fontsize=10, fontweight='bold', color='blue')
plt.show()
```

### 애니메이션 효과 사용하기

```python
filename = './전_산업의_자격증별_데이터_전문_인력_우대사항_20240429180042.csv'
happiness = pd.read_csv(filename, encoding='cp949')
happiness.head()
```

```python
# 애니메이션 효과
fig = px.sunburst(data_frame=happiness,
                  path=['산업별', '자격증별'],
                  values='입사우대',
                  color='입사우대없음')
fig.update_layout(title='자격증 입사 우대 자격증 지수', title_x=0.5, width=1200, height=900)
fig.show()
```
